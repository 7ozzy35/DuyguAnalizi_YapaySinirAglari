### 2. regex(regular expressions) kullanarak HTML taglerini temizleme

```python
def remove_tags(raw_text):
    cleaned_text = re.sub(re.compile('<.*?>'), "", raw_text)
    return cleaned_text
```

```python
data['review'] = data['review'].apply(remove_tags)
data['review'][0]
```

### 3. Küçük harfe dönüştürme

```python
data['review'] = data['review'].apply(lambda x:x.lower())
data['review'][0]
```

### 4.Noktalama işaretlerini kaldırma

```python
import string
removed = string.punctuation
def remove_punctuation(text):
    for r in removed:
        text = text.replace(r,'')
    return text
```

```python
data['review'] = data['review'].apply(remove_punctuation)
data['review'][0]
```

```python
import nltk
nltk.download('stopwords')
```

### 5. Belirli 'stopwords' leri kaldırma

```python
from nltk.corpus import stopwords

word_list = stopwords.words('english')
data['review'] = data['review'].apply(lambda x: [word for word in x.split() if word not in word_list]).apply(lambda x:" ".join(x))
```

## Tokenization

```python
# Bazı özellik değerleri önceden tanımlanır ki sonraki aşamalarda değişiklik yapılmak istenirse daha kolay olsun.
vocab_size = 50000
max_length = 2000
trunc_type='post'
padding_type='post'
oov_tok = ""
```

```python
X_train.shape
X_train = X_train['review'].values
X_train.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok) # Tokenizer nesnesi oluşturulur.num_words=vocab_size, oov_token=oov_tok
tokenizer.fit_on_texts(X_train) # training veri setine uygulanır.
```

```python
word_index = tokenizer.word_index

training_sequences = tokenizer.texts_to_sequences(X_train)
training_padded = pad_sequences(training_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

testing_sequences = tokenizer.texts_to_sequences(X_test)
testing_padded = pad_sequences(testing_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
embedding_dim = round(vocab_size ** 0.25) # use the fourth root of the vocab size
l2_regularization = 0.01
```

```python
import tensorflow as tf

def create_model(vocab_size, embedding_dim, l2_regularization):
    model = tf.keras.Sequential([
            tf.keras.layers.Embedding(vocab_size, embedding_dim), # turns positive integers into dense vectors of fixed size
            tf.keras.layers.GlobalAveragePooling1D(),
            tf.keras.layers.Dense(24, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(l2_regularization)),
            tf.keras.layers.Dropout(.25),
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])

    adam = tf.keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=False)
    
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

    return model
```

```python
model = create_model(vocab_size, embedding_dim,l2_regularization)
# Display the model's architecture
model.summary()
```
